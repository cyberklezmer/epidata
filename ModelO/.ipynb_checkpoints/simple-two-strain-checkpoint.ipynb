# Model fitting

Model:
$$
X_{t} = X_{t-1}  I_{t-1} s_{t-1} c_{t-2} r_{t} + \epsilon_t
$$
where

$s_{t-1}$ is seasonal week ago

$c_{t-2}$ is contacts reduction two weeks ago 

$r_{t}$ is factor depending on the virus strain

$I_{t-1}$ is combined effect of vaccination and prior infection

$$
I_t = (1-u_t)(1-v_t - w_t),
$$
where, roughly, 

$u_t$ is percentage protected by infection (ascertainment = 0.4)

$v_t$ is percentage protected by second dose

$w_t$ is percentage protected by boosters

```python
from sklearn.linear_model import LinearRegression

def get_I(df, alpha=ALPHA, omicron=OMICRON_DATA):
    x = (df.X/POPULATION).to_numpy()
    f = (df.F/POPULATION).to_numpy()
    b = (df.B/POPULATION).to_numpy()
    y = (df.Y/POPULATION).to_numpy()
    
    ud = np.convolve(rho, x + omicron['immunity_escape_inf']*y)
    uo = np.convolve(rho, omicron['immunity_escape_inf']*x + y)
#     v = np.convolve(phi, f)

    ff = np.zeros((len(ud), len(ud)))
    for tau in range(len(f)):
        for t in range(len(f)):
            if t < tau + 30: # not eligible for booster yet
                ff[tau, t] = f[tau]
            elif t >= tau + 30: # some people got boosted after 5 months and we are recording those who did not get boosters
                ff[tau, t] = f[tau] - b[tau+30]            
                
    v = np.zeros(len(ud))
    for t in range(1, len(ud)):
        v[t] = np.convolve(phi, ff[:, t])[t]                
    

#     for t in range(1,len(v)-30):
#         v[t] = np.convolve(phi, f[tau] - b[30:t+30])[t]
        
    w = np.convolve(psi, b)
    
    I = (1-ud/alpha)*(1-v-w)
    
    J = (1-uo/alpha)*(1-omicron['immunity_escape_vac']*v-omicron['immunity_escape_boost']*w)
    
    ret_df = pd.DataFrame({'I': I[:len(df)],
                           'ud': ud[:len(df)],
                           'uo': uo[:len(df)],
                           'v': v[:len(df)],
                           'w': w[:len(df)],
                           'J': J[:len(df)],
                           'x': x[:len(df)],
                           'y': y[:len(df)], 
                        }, index=df.index)
        
    return ret_df

def get_r(df, alpha=ALPHA, plot=False): # calculates r_t for the delta strain
#     df = get_epidemy_df()
#     df = df[df.index <= pd.Timestamp(2021, 10, 15)]
    I = get_I(df, alpha=alpha)
    I = I[I.index >= DELTA_DATE]['I'].to_numpy()

    c = df[df.index >= (DELTA_DATE - pd.Timedelta('1W'))]['c'].to_numpy()[:-1] # we need c_{t-2}
    # print(c)

    df = df[df.index >= DELTA_DATE]
    # print(df.X)
    X = df.X.to_numpy()
    y = X[1:]/X[:-1] # X_t / X_{t-1}
    x = I[:-1] * c[:-1] * df.s.to_numpy()[:-1]
#     r = np.dot(y,x)/np.dot(x,x)
    # s = (np.dot(y,y) - r*np.dot(x,y))/(len(x) - 1)
#     print(r, s)

    reg = LinearRegression(fit_intercept=False).fit(x.reshape(-1, 1), y)
#     print(reg.coef_, reg.intercept_)
    r = reg.coef_[0]
#     print(reg.score(x.reshape(-1, 1), y))


#     rs = y/x
#     print(rs.min(), rs.max())

    if plot:

        fig, ax = plt.subplots(1, 1, figsize=(16,8))
        plt.plot(x, y, '.')
        t = np.arange(x.min(), x.max(), 0.01)


        plt.plot(t, r*t)
        # t = np.arange(0, 0.5, 0.01)
        # plt.plot(t, rs.min()*t)
        # plt.plot(t, rs.max()*t)
        # a = sp.stats.t.ppf(1-0.05, len(x)-1)*s/np.sqrt(np.dot(x,x))
        # plt.fill_between(t, (b+a)*t, (b-a)*t, alpha=0.3)
        # plt.xlim([0,0.5])
        # plt.ylim([0,3])

    #     plt.plot(y/r*x, '.')

    return r
    
    

def get_r_alpha(df): # calculates r_t and alpha for the delta strain via ordinary least squares 
    c = df[df.index >= (DELTA_DATE - pd.Timedelta('1W'))]['c'].to_numpy()[:-1] # we need c_{t-2}
    s = df[df.index >= DELTA_DATE]['s']
#     print(s)
    # print(c)

    # print(df.X)
    X = df[df.index >= DELTA_DATE]['X'].to_numpy()
    y = X[1:]/X[:-1] # X_t / X_{t-1}

    dfI = get_I(df)
    dfI = dfI[dfI.index >= DELTA_DATE]
    
    # X_t ~ I_{t-1}
    x1 = (1-dfI['v']-dfI['w']).to_numpy()[:-1]*c[:-1]*s[:-1]
    x2 = (dfI['ud']*(1-dfI['v']-dfI['w'])).to_numpy()[:-1]*c[:-1]*s[:-1]
    
    # X_t ~ I_t # actually performs worse
#     x1 = (1-dfI['v']-dfI['w']).to_numpy()[1:]*c[:-1]*s[:-1]
#     x2 = (dfI['ud']*(1-dfI['v']-dfI['w'])).to_numpy()[1:]*c[:-1]*s[:-1]
    
    x = np.vstack([x1, x2]).T
    
    reg = LinearRegression(fit_intercept=False).fit(x, y)
#     print(reg.coef_, reg.intercept_)
#     print(reg.score(x, y))
    r = reg.coef_[0]
    alpha = -r/reg.coef_[1]
    return r, alpha

```

```python
def predict(n, 
#             df,
            alpha=ALPHA,
            omicron=OMICRON_DATA,
           ):
    df = get_epidemy_df()
    
    # first we extend df[['c', F', 'B']] by constants and take the true seasonal
    last_date = df.index[-1]
    week = pd.Timedelta('1W')
    res = pd.DataFrame([df.iloc[-1]]*n)
    res.index = [last_date + i*week for i in range(1, n+1)]
    res['X'] = np.nan
#     res['B'] = np.zeros(n)
#     res['F'] = np.zeros(n)
    res['s'] = seasonal.loc[res.index]
    new_df = df.append(res)
    

    # import omicron
    new_df['Y'] = np.zeros(len(new_df))
    new_df.loc[last_date]['Y'] = omicron['input_size']

    # let's estimate r
    r = get_r(df, alpha=alpha, plot=False)
    
    # finally we predict X recalculating the whole I-table for each new prediction 
    for k in range(-n, 0):
        dfI = get_I(new_df.iloc[:k], alpha=alpha, omicron=omicron)
        tmp = new_df.iloc
        new_df.iloc[k]['Y'] = tmp[k-1]['Y']*tmp[k-1]['s']*tmp[k-2]['c']*dfI.iloc[-1]['J']*r*omicron['rate_change']
        print(new_df.iloc[k]['Y'])
        new_df.iloc[k]['X'] = tmp[k-1]['X']*tmp[k-1]['s']*tmp[k-2]['c']*dfI.iloc[-1]['I']*r
    return new_df
```

```python
import holoviews as hv
from holoviews import opts

hv.extension('bokeh')
```

```python
def calculate_omicron_r(omicron=OMICRON_DATA): # calculates the omicron rate from doubling time based on UK data
    J = (1 - omicron['immunity_escape_inf']*0.5)*(1-omicron['immunity_escape_vac']*0.3-omicron['immunity_escape_boost']*0.5)
#     return np.exp(7*np.log(2)/omicron['doubling_time'])/J
    return np.exp(7*(np.log(2) - np.log(J))/omicron['doubling_time'])

def plot_prediction(**args):
    omicron = OMICRON_DATA.copy()
    omicron.update(args)
    n = 12
    omicron['input_size'] = 3500
    omicron['rate_change'] = calculate_omicron_r(omicron)
#     print(omicron['rate_change'])
    print(omicron)
    
    
    df = predict(n, alpha=alpha, omicron=omicron)
    df = df[df.index >= pd.Timestamp(2021, 11, 1)]
    delta = hv.Scatter(df, ('index', 'Datum'), ('X', 'Cases')).opts(framewise=True)
    omicron = hv.Scatter(df, ('index', 'Datum'), ('Y', 'Cases')).opts(framewise=True)
    df['both'] = df['X'] + df['Y']
    both = hv.Curve(df, ('index', 'Datum'), ('both', 'Cases')).opts(framewise=True, color='green')
    return delta * omicron * both
```
